```python
import pandas as pd
```

```python
df = pd.read_csv('/content/iris.csv')
```

```python
df
```

```python
X = df.drop(columns=['species'])
y = df['species']
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical  # একসময়েই import করলাম যেই জায়গায় লাগবে

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
y_one_hot = to_categorical(y_encoded)
```

```python
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_one_hot, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

model = Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(10, activation='relu'),
    layers.Dense(y_one_hot.shape[1], activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=50, batch_size=8, validation_split=0.1)
```

```python
y_pred = model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

y_test_labels = y_test.argmax(axis=1)
y_pred_labels = y_pred.argmax(axis=1)

print(classification_report(y_test_labels, y_pred_labels))
print(confusion_matrix(y_test_labels, y_pred_labels))
```
